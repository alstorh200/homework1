```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import imshow
```

```python
img = cv2.imread('mm.jpg')
if img is None:
    print('Could not read image!')
else:
    print('ok')
```

```python
plt.imshow(img)
```

```python
print(img)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title('image after correction')
```

Assignment_1

```python
img_rgb = img[:,:,::-1]
plt.imshow(img_rgb)
plt.title('image after correction')
```

```python
print('Dimensions: ', img_rgb.shape)
print('Data type: ', img_rgb.dtype)
print('Min Pixel: ', img_rgb.min())
print('Max Pixel: ', img_rgb.max())
```

```python
pixel = img_rgb[100,150]
print('pixel value: ', pixel)
```

```python
img_modified = img_rgb.copy()
img_modified[50:120 , 80:160] =[255,0,0]
plt.imshow(img_modified)
plt.title(' after edit pixel')
plt.axis('off')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
```

```python
plt.hist(gray.ravel(), bins=256 , color='black')
plt.title('')
plt.xlabel('Intensity')
plt.ylabel('Count')
```
