```python
from math import gamma

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_bgr =plt.imread("mm.jpg")
img_bgr = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_bgr)
plt.title('Original Image')
plt.axis("off")
```

```python
gray = cv2.cvtColor(img_bgr, cv2.COLOR_RGB2GRAY)
plt.imshow(gray , cmap='gray')
plt.title('Gray Image')
plt.axis("off")
```

```python
plt.hist(gray.ravel(),bins=256)
plt.title('Histogram')
plt.xlabel(' ')
plt.ylabel(' ')
plt.show()
```

Histogram Equalization

```python
he = cv2.equalizeHist(gray)
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
plt.imshow(gray , cmap='gray')
plt.title(' ')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(he, cmap='gray')
plt.title('Histogram Equalization')
plt.axis('off')
plt.show()
```

Contrast Stretching

```python
cs = cv2.normalize(gray, None, 0, 255, cv2.NORM_MINMAX)
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
plt.imshow(gray , cmap='gray')
plt.title(' ')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(cs, cmap='gray')
plt.title('Contrast Stretching')
plt.axis('off')
plt.show()
```

Gamma Correction

```python
gamma = 0.6
gamma_img = np.uint8(np.power(gray/255.0,gamma)*255)
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title(' ')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(gamma_img, cmap='gray')
plt.title(f"Gamma Correction (Y={gamma})")
plt.axis('off')
plt.show()
```

CLAHE --Adaptive Histogram Equalization

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
clahe_img = clahe.apply(gray)
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title('')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(clahe_img, cmap='gray')
plt.title('CLAHE')
plt.axis('off')
plt.show()
```
