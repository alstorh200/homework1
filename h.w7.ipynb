```python
import cv2
import numpy as np

img = cv2.imread("mm.jpg")
img = cv2.resize(img, None, fx=0.7, fy=0.7)

# تحويل إلى YCrCb
ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)

# حدود لون الجلد
lower = np.array([0, 133, 77], dtype=np.uint8)
upper = np.array([255, 173, 127], dtype=np.uint8)

skin_mask = cv2.inRange(ycrcb, lower, upper)

# تنعيم الماسك
skin_mask = cv2.GaussianBlur(skin_mask, (7, 7), 0)
```

```python
# تنعيم يحافظ على الحواف
smooth_skin = cv2.bilateralFilter(
    img,
    d=9,
    sigmaColor=75,
    sigmaSpace=75
)
```

```python
skin_mask_3c = cv2.merge([skin_mask, skin_mask, skin_mask]) / 255.0

result = (smooth_skin * skin_mask_3c + img * (1 - skin_mask_3c)).astype(np.uint8)

cv2.imshow("Original", img)
cv2.imshow("Skin Mask", skin_mask)
cv2.imshow("Beauty Result", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
skin_mask = cv2.morphologyEx(skin_mask, cv2.MORPH_CLOSE, kernel)
```

```python
alpha = 0.6
result = cv2.addWeighted(result, alpha, img, 1-alpha, 0)
```
