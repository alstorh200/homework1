```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("mm.jpg")
img= cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(img); plt.title("  "); plt.axis('off')
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
kernel
```

```python
eroded = cv2.erode(img, kernel, iterations=1)
plt.Figure(figsize=(10,5))
plt.Subplot(1,2,1); plt.imshow(img); plt.title('original image'); plt.axis('off')
plt.Subplot(1,2,2); plt.imshow(eroded); plt.title('Erosion'); plt.axis('off')
```

```python
dilated = cv2.dilate(img , kernel iterations=1)
plt.Figure(figsize=(10,5))
plt.Subplot(1,2,1); plt.imshow(img); plt.title('original image'); plt.axis('off')
plt.Subplot(1,2,2); plt.imshow(dilated); plt.title('dilated'); plt.axis('off')
```

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
plt.Figure(figsize=(10,5))
plt.Subplot(1,2,1); plt.imshow(img); plt.title('original image'); plt.axis('off')
plt.Subplot(1,2,2); plt.imshow(opening); plt.title('opening'); plt.axis('off')
```

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
plt.Figure(figsize=(10,5))
plt.Subplot(1,2,1); plt.imshow(img); plt.title('original image'); plt.axis('off')
plt.Subplot(1,2,2); plt.imshow(closing); plt.title('closing'); plt.axis('off')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
edges = cv2.Canny(gray,80,160)
plt.imshow(edges, cmap='gray'); plt.title('Edge Mask   '); plt.axis('off')
```

```python
soblel_x = cv2.Sobel(gray, cv2.CV_64F,1,0,kernel=3)
soblel_y = cv2.Sobel(gray, cv2.CV_64F,0,1,kernel=3)
solbl = np.sqrt(soblel_x**2 + soblel_y**2)
edges2 = np.uint8(solbl)
plt.imshow(edges2, cmap='gray'); plt.title('Edge mask   '); plt .axis('off')
```

```python
edges_disted = cv2.dilate(edges,kernel)
edges_clean = cv2.morphologyEx7(edges_disted, cv2.MORPH_CLOSE, kernel)
plt.Figure(figsize=(12,6))
plt.Subplot(1,3,1); plt.imshow(edges_clean, cmap='gray'); plt.title('Dislated Edges') ; plt.axis('off')
plt.Subplot(1,3,2); plt.imshow(edges, cmap='gray'); plt.title('edges_clean') ; plt.axis('off')
```
